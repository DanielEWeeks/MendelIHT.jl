# Create plots for precision/recall curves

```julia
using Plots
using DelimitedFiles
```

```julia
myruns = collect(1:30)

iht_precision_normal   = readdlm("./normal_results/iht_precision_normal")
lasso_precision_normal = readdlm("./normal_results/lasso_precision_normal")
iht_recall_normal      = readdlm("./normal_results/iht_recall_normal")
lasso_recall_normal    = readdlm("./normal_results/lasso_recall_normal")
normal_precision = [iht_precision_normal lasso_precision_normal]
normal_recall = [iht_recall_normal lasso_recall_normal];
```

```julia
normal_precision_plot = plot(myruns, normal_precision, title = "Normal Precision Plot", 
    xlabel = "Runs", ylabel = "Precision", w=3, lab = ["IHT" "LASSO"], ylim = (0, 1.01),
    legend=:top, linestyle = [:solid :dash])

normal_recall_plot = plot(myruns, normal_recall, title = "Normal Recall Plot", 
    xlabel = "Runs", ylabel = "Recall", w=3, lab = ["IHT" "LASSO"], ylim = (0, 1.01),
    legend=:bottomright, linestyle = [:solid :dash])

normal_precision_recall = [normal_precision_plot normal_recall_plot]
normal_plot = plot(normal_precision_recall..., layout = (1, 2),size = (800, 300))
```

```julia
normal_plot = plot(normal_precision_recall..., layout = (1, 2),size = (800, 300), dpi=300)
savefig(normal_plot, "normal_plot")
```
