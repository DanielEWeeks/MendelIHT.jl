# IHT.jl tutorial (brief)

In this tutorial we explore some of the functionality of the IHT.jl package, which implements iterative hard thresholding on floating point arrays and binary PLINK data. 

The first part of the tutorial demonstrates how to handle floating point data. Later we will show how to use [PLINK binary genotype files](http://pngu.mgh.harvard.edu/~purcell/plink/binary.shtml). IHT.jl offers two parallel computational frameworks for PLINK data: multicore CPUs or massively parallel GPUs. Although this tutorial features the CPU version, instructions for GPU use are included.

IHT performs penalized least squares regression with the loss function known as the _residual sum of squares_ $\frac{1}{2} \| \boldsymbol{y} - \boldsymbol{X} \boldsymbol{\beta} \|_2^2 = \sum_{i = 1}^{n} (y_i - \boldsymbol{x}_i^T \boldsymbol{\beta})^2$ for data $\boldsymbol{X}$, response vector $\boldsymbol{y}$, and statistical model $\boldsymbol{\beta}$. In regression settings with more predictors than cases, IHT chooses a parsimonious sparse statistical model $\boldsymbol{\beta}$ with a user-specified number of nonzeroes.

This tutorial targets novice users. Advanced users are encouraged to read the full tutorial at `~/.julia/v0.4/IHT/IHT_tutorial_full.ipynb`.

Before starting, ensure that the necessary packages are installed. This tutorial uses IHT.jl, PLINK.jl and DataFrames.jl. They can be installed from the Julia command line via the commands

    Pkg.clone("https://github.com/klkeys/PLINK.jl") # unregistered packages must be cloned
    Pkg.clone("https://github.com/klkeys/IHT.jl")
    Pkg.add("DataFrames") # registered packages installed with different interface

We begin by adding processors and loading libraries. Here we use the internal Julia variable `CPU_CORES` to enlist all available CPU cores for our calculations. Unfortunately, you may see many benign warnings when loading these packages. They will disappear in Julia v0.5.

```julia
addprocs(CPU_CORES) # important for parallel crossvalidation
using IHT           # module IHT.jl performs analysis
using PLINK         # module PLINK.jl handles PLINK data
using DataFrames    # pretty printing of results
```

## IHT for numeric data

For this tutorial, we will use a script called `tutorial.simulation.jl` to simulate all data and variables. The script is located at `simpath`:

```julia
simpath = "~/.julia/v0.4/IHT/sim/tutorial_simulation.jl"
simpath = expanduser(simpath) # unravels the "~" in simpath
include(simpath) # simulate the data
```

### Loading numeric data

The script `tutorial_simulation.jl` generates a floating point data matrix `X` with a response vector `y`. The vector `y` is simulated from a sparse statistical model $\boldsymbol{\beta}$ with $k_{\text{true}} = 10$ nonzero coefficients. In practice, users will load their own data from file using the commands

    X = readdlm(PATH_TO_X.txt)      # load data matrix X
    y = vec(readdlm(PATH_TO_Y.txt)) # read response y and vectorize it

The Julia function `readdlm` **read**s **d**e**l**i**m**ited files such as CSV or tab-delimited TXT into a matrix. It offers several optional keyword arguments to correctly handle the variety of possible delimited files. If the delimited file includes a header, then `X` is read via the call

    (X, h) = readdlm(PATH_TO_X_WITH_HEADER.txt, header=true)

Here the array `h` contains the header names. Another optional argument `delim` specifies the text **delim**iter. The default delimiter is whitespace, but `delim` accepts any suitable ASCII character as a delimiter. If data are stored in CSV, then use `readdlm` or just `readcsv`:

    X = readdlm(PATH_TO_X.csv, ',') # read delimited file with comma delimiter
    X = readcsv(PATH_TO_X.csv)      # preferred CSV read command

### Running IHT with numeric data

With our data `X` in hand, we now run iterative hard thresholding to capture the best model of size `k = 10`. The function call `L0_reg` returns an object `output` with the following fields:

- `output.time`, the compute time in seconds
- `output.iter`, the number of iterations taken until convergence
- `output.loss`, the residual sum of squares at convergence
- `output.beta`, the $\boldsymbol{\beta}$ vector at convergence

```julia
output = L0_reg(X,y,k)  # predictors X, response y, and model size k
```

Let us compare the results to the true model `β`:

```julia
bk = copy(output.beta)  # estimated coefficients β
DataFrame(True_β=b[bidx], # compare true β to estimated β
    Estimated_β=bk[bidx]) # bidx indexes true model
```

We see that IHT returns all the true nonzero coefficients. The coefficient values themselves are fairly close to their originals. We expect this since the standard deviation `s = 0.1` of the noise level of the simulated data does not yield a very noisy `y`. Observe what happens when we use a noisier response `y2` with `s = 5.0` simulated from the same data `X`:

```julia
output2 = L0_reg(X, y2, k)   # run IHT with noisier response y2
bk2     = copy(output2.beta) # copy results for later use
DataFrame(True_β=b[bidx], Noisy_β=bk[bidx], Noisier_β=bk2[bidx])
```

The table compares our analysis of `y` (column **Noisy**) and `y2` (column **Noisier**) against the true simulated response (column **True**). The coefficient values for the noisier response `y2` are markedly less accurate, and several nonzero values are not recovered. The failure to recover the correct model does not indicate that IHT model selection performance is necessarily bad; the high noise level makes accurate estimation of the model quite difficult.

## IHT for GWAS

IHT.jl interfaces with [PLINK.jl](https://github.com/klkeys/PLINK.jl) for GWAS analysis. Although PLINK.jl primarily handles PLINK binary genotype files (BED files), it also requires the corresponding BIM and FAM files to be in the same directory. In this section of the tutorial we will demonstrate GWAS analysis with IHT.jl. 

### Loading genotype data

For the purpose of the tutorial, we will use simulated binary genotype from PLINK.jl.
Our binary genotype data are housed in `xbed` and the corresponding simulated response is `ybed`.
To instead load genotype data and the phenotype from the PLINK data files stored at `xpath`, we use the command

    (xbed, ybed) = read_plink_data(xpath)

Note that `xpath` obeys the PLINK convention of naming the three data files the same. In other words,

    (xbed, ybed) = read_plink_data("mydata")

reads the files `mydata.bed`, `mydata.bim`, and `mydata.fam` in the local directory.
This call to `read_plink_data` initialized `xbed` with an intercept (column of $1$s) by default.
The phenotype `ybed` is read as the rightmost column of the FAM file.
For phenotypes stored elsewhere, the call

    (xbed, ybed) = read_plink_data(xpath, ypath)

reads `ybed` from `ypath` instead. Here PLINK.jl assumes that `ypath` points to a delimited text file, so the previous discussion about `readdlm` applies here. If the file at `ypath` contains a header, then parse the header with the optional keyword argument `header = true`. To change the delimiter, use the optional keyword argument `delim`. The default assumes no header and a whitespace delimiter.

To include nongenetic covariates stored in a delimited text file at `covpath`, then call

    (xbed, ybed) = read_plink_data(xpath, covpath, ypath)

Here the intercept is **not** added by default, as PLINK.jl assumes that it is stored in `covpath`.
PLINK.jl expects that both `ypath` and `covpath` point to delimited text files.
Control the headers and delimiters with the optional keyword arguments

- `yheader` for the header to the phenotype file; default is `false`.
- `covheader` for the header to the covariate file; default is also `false`.
- `ydelim` for the delimiter to the phenotype file; default is whitespace.
- `covdelim` for the delimiter to the covariate file; default is whitespace.

For example, to load `ybed` from a text file with no header and the covariates from a CSV file with a header, call

    (xbed, ybed) = read_plink_data(xpath, covpath, ypath, 
                                   covheader = true, covdelim = ',')

Note that IHT **does not** work with categorical covariates! All covariates must either be numerical or quantitative $0/1$ variables. Furthermore, IHT **WILL NOT** work with missing data! Any missing genotypes, phenotypes, or covariates must be suitably imputed before running IHT.

### IHT with GWAS data

For analysis with PLINK files, the call to `L0_reg` is the same as with the floating point `X`:

```julia
output = L0_reg(xbed, ybed, k) # run IHT with BED files
```

As before, let us compare the estimated model to the simulated one:

```julia
bk3 = copy(output.beta) # copy the beta for later use
DataFrame(β=bbed[bidxbed], Estimated=bk3[bidxbed])
```

As expected, the coefficient estimates are reasonably accurate.

## Crossvalidation with numeric data

In a realistic setting, we do not know the correct model size $k_{\text{true}} \,$. IHT solves this problem by [crossvalidating](https://en.wikipedia.org/wiki/Cross-validation_%28statistics%29) the best model size $k_{\text{best}} \,$ with the function `cv_iht`.

Crossvalidation is a standard statistical procedure for performing model validation in the absence of a validation data set,that is, a second data set independent of our data $\boldsymbol{X}$. The data $\boldsymbol{X}$ are split into $q$ mutually exclusive subsets called _folds_. Then $\boldsymbol{\beta}$ is estimated on the subset of $\boldsymbol{X}$ containing $q - 1$ folds and independently validated on the $q$th fold (sometimes called the _holdout set_) with some crossvalidation criterion. Each of the $q$ folds takes a turn as a hold-out set. At the end, the results are averaged across all $q$ folds.

We use [mean squared error (MSE)](https://en.wikipedia.org/wiki/Mean_squared_error) as our crossvalidation criterion. MSE is merely the residual sum of squares on the $q$th fold.

Let us use `cv_iht` to compute $k_{\text{best}} \,$ for the very noisy response `y2` with MSE as our crossvalidation criterion:

```julia
srand(2016) # reset seed before crossvalidation for reproducible results
cv_output = cv_iht(X, y2)
```

IHT calculates an ideal model size $k_{\text{best}} = 6$ that is smaller than the true model size $k_{\text{true}} = 10$, again highlighting the negative effects of noise on our ability to correctly identify the true statistical model.

The function `cv_iht` sets reasonable defaults for the crossvalidation fold structure, the models to test, and the desired number of crossvalidation folds. Users can modify these with the following optional keyword arguments:

- `q` is the number of folds to compute. The default depends on the number of available CPU cores, but it usually defaults to 5. An error is thrown if `q <= 2`.
- `path` is an integer vector of $N$ model sizes $k_1, k_2, \ldots, k_N$ to test. The default is to test sizes $1, 2, \ldots, 20$.
- `folds` is an integer vector that assigns data to each fold. The default is to randomly assign data as evenly as possible to each fold.

IHT provides a `plot` function to generate a simple line plot of MSE versus model size from crossvalidation results:

```julia
IHT.plot(cv_output)
```

This default plotting function adds a red line at $k_{\text{best}}$. Users who wish to generate their own plots and are comfortable with the Julia command line have several plotting tools available to them. Some notable ones include:

- [Gadfly.jl](http://gadflyjl.org/) (IHT uses Gadfly for plotting)
- [Plots.jl](https://juliaplots.github.io/)
- [PyPlot.jl](https://github.com/stevengj/PyPlot.jl)

Alternatively, users can write the MSEs and model sizes to file and then load them in their favorite plotting environment, such as [R](https://www.r-project.org/). One simple way to save MSEs and model sizes to file is to call `readdlm`'s companion function `writedlm`:

    writedlm(PATH_TO_NEW_FILE.txt, [x.path x.mses])

This call creates the matrix `[x.path x.mses]` and saves it to file.

## Crossvalidation with genotype data

IHT.jl can use PLINK files directly in parallel crossvalidation. However, the interface of `cv_iht` changes in two crucial ways. First, instead of feeding the data `X` and `y` directly, we must point `cv_iht` to the locations of all data files stored on disk. This means that `cv_iht` will **not** read PLINK files loaded with `read_plink_data`! Second, all data except the covariates must be stored in *binary* format.

Storing data in binary format is actually quite simple. The call in Julia is `write`. For example, the script `tutorial_simulation.jl` saved the response variable `ybed` to a binary file `y.bin` at `ypath` by calling

    write(open(ypath, "w"), ybed)
 
Our simulation script generated the correct filepaths `xpath`, `covpath`, and `ypath` to simulated genotype data stored in the PLINK.jl module. Here `xpath` points to the PLINK files, `covpath` points to the delimited text file containing covariates, and `ypath` points to the response vector in a delimited text file.

Now let us run `cv_iht` with binary data:

```julia
srand(2016) # reset seed to make reproducible results
cv_bed  = cv_iht(xpath, covpath, ypath)
```

We see that IHT finds the correct model size $k_{\text{true}} = 10$. IHT also captures the correct predictors and provides reasonably accurate estimates of the coefficients:

```julia
DataFrame(β_idx = bidxbed, cv_idx = cv_bed.bidx,
          True_β = bbed[bidxbed], Estimated_β = cv_bed.b)
```

We can also plot MSEs versus model size:

```julia
IHT.plot(cv_bed)
```

## GPUs

As mentioned before, IHT.jl can use GPUs to accelerate computations. This is an advanced topic, so we will only outline its use. We will not execute any code here since we cannot assume that the user has a dedicated GPU on their machine.

Using GPUs with IHT.jl require both a `BEDFile` object and a string object containing the GPU code. PLINK.jl preloads the requisite GPU code into the variable `PLINK.gpucode64`. Assuming that a suitable GPU device is available on the machine, one can call `L0_reg` with GPU facilities using

    output = L0_reg(xbed, ybed, k, PLINK.gpucode64)

Using GPUs in crossvalidation merely entails adding `PLINK.gpucode64` as an argument to `cv_iht`:

    cv_bed = cv_iht(xpath, covpath, ypath, PLINK.gpucode64)

For more involved help with the GPU code, [email](mailto:klkeys@g.ucla.edu) the IHT.jl developers.
